# Clustering

To further understand the dataset, especially the text column 'description', this notebook does two main things:

1. take K-Means Clustering to analyze 'Release Year' and 'Durati''n
2. take K-Means Clustering to analyze the distribution of the five main topics of 'descripton'o.

## dataset import

```python
import pandas as pd
import numpy as np
import os
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset/original/netflix_titles.csv')
df.head(3)
```

## dataset prepare

```python
# Dealing with missing director information
df['director'].replace(np.nan, 'No Data',inplace  = True)

# Drops all lines containing missing values
df.dropna(inplace=True)

# Drop Duplicates lines
df.drop_duplicates(inplace= True)

# Filter the subset that contains only movie information
df_movies = df.loc[df['type'] == 'Movie']

# Assuming the 'duration' column is in the format '125 min'
df_movies['duration'] = df_movies['duration'].str.extract('(\d+)', expand=False)

# Convert the 'duration' column to numeric '125'
df_movies['duration'] = pd.to_numeric(df_movies['duration'], errors='coerce')

# verify dataframe
print(df_movies.isnull().sum())
print(df_movies.info())
```

## K-Means Clustering of 'Release Year' and 'Duration'

It's more like a map of the distribution of movies of different durations across generations

```python
import seaborn as sns

# Select the columns for clustering
X = df_movies[['release_year', 'duration']]

# Specify the number of clusters
n_clusters = 3

# Create a KMeans instance
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the data
df_movies['cluster'] = kmeans.fit_predict(X)

# Specify a custom color palette
custom_palette = sns.color_palette("husl", n_clusters)
# Visualize the clusters
sns.scatterplot(x='release_year', y='duration', hue='cluster', data=df_movies, palette=custom_palette)
plt.title('K-Means Clustering of Release Year and Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.show()
```

## K-Means Clustering of Movie Theme(Description)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create a TfidfVectorizer with English stop words
vectorizer = TfidfVectorizer(stop_words='english')
# Transform the 'description' column using TfidfVectorizer
X = vectorizer.fit_transform(df_movies['description'])

## Create a KMeans instance
num_clusters = 5 
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the model to the vectorized data
df_movies['cluster'] = kmeans.fit_predict(X)

# Display the entire 'description' column without truncation
pd.set_option('display.max_colwidth', None)
for i in range(num_clusters):
    print(f"Cluster {i}:\n")
    print(df_movies[df_movies['cluster'] == i]['description'].head(10))
    print("\n")
```

Based on the clustering results provided, we can initially summarize the general themes of each cluster:

Cluster 0.
This cluster seems to contain a number of movies related to thrillers, crime, espionage and action. The descriptions involve plots such as models returning to historical slavery, terrorist attacks, blackmail and shark attacks.

Cluster 1.
This cluster seems to involve some movies that deal with family, romance, humor, and a lighthearted atmosphere. Descriptions include plots dealing with loss, strange birds, and twins getting married.

Cluster 2.
This cluster seems to contain some movies that relate to science fiction, adventure, war, and supernatural elements. Descriptions include plots dealing with alien invasions, fleeing from war, supernatural forces, and crime epics.

Cluster 3.
This cluster seems to contain a number of movies related to friendship, comedy, youth, and memories. Descriptions include plots such as middle-aged friend reunions, YouTube celebrities, and friendship adventures.

Cluster 4.
This cluster seems to contain some movies related to romance, war, family, and love stories. Descriptions involve plots such as romantic poets, war love, desert island romance, etc.

```python
plt.figure(figsize=(10, 6))
for i in range(num_clusters):
    cluster_data = df_movies[df_movies['cluster'] == i]
    plt.scatter(cluster_data['release_year'], cluster_data['duration'], label=f'Cluster {i}')

plt.title('Movie Theme Clusters')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.legend()
plt.show()
```
